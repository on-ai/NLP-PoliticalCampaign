```python
# Function scraping twitter until user presses stop in Jupyter

def twint_scrap (sentence, output_file):
    ''' Function scraping twitter until user presses "interrupt the kernel" in Jupyter, based on the twint package
    Arguments are:
    - sentence to scrap
    - file to generate the output to
    
    Usage: twint_scrap("Fuck la police", "c:\\temp\\covid.csv")
    '''
#     import twint
    
    # Configure
    nest_asyncio.apply()
    c = twint.Config()
    c.Search = sentence
    c.Store_csv = True

    c.Output = output_file
    twint.run.Search(c)
    
    return

```

```python
lat_list = []
lon_list = []
for point in data_places['place']:
    point = point.replace(r"[", r"(")
    point = point.replace(r"]", r")")
    for match in re.findall(r'(?<=\().*?(?=\))',point):
        new_point_list.append(match)
        a,b = map(float,match.split(','))
        lat_list.append(a)
        lon_list.append(b)
```

```python
m = folium.Map(location=[-34.2161989, -70.80328209],tiles='cartodbpositron',zoom_start=11)
for idx in range(len(lat_list)):
    folium.Marker(location=[lat_list[idx], lon_list[idx]],icon=folium.Icon(color='red',icon='none')).add_to(m)
display(m)
```
